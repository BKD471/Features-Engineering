# One Hot encoding for features with lots of categories

```python
import numpy as np
import pandas as pd
data=pd.read_csv('train.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
data.columns.unique()
```

```python
data.columns.value_counts()
```

```python
for col in data.columns:
    print(col,' :',len(data[col].unique()),'labels')
```

```python
pd.get_dummies(data,drop_first=True).shape
```

for variable of 6 we get 117 additional new features

```python
# top 10 most frequent categories for variable x2

data.X2.value_counts().sort_values(ascending=False).head(10)
```

```python
top10=[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top10
```

encoding top10 most frequent categories for x2 and appending them to the table

```python
for label in top10:
    data[label]=np.where(data['X2']==label,1,0)
    
    
```

```python
data
```

```python
def one_hot_top(data,label,top):
    
    for x in top:
        data[label+'_'+x]=np.where(data[label]==x,1,0)
    
    
data=pd.read_csv('train.csv',usecols=['X1','X2','X3','X4','X5','X6'])
one_hot_top(data,'X2',top10)
```
